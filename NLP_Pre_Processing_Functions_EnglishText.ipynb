```python
import nltk
nltk.download('all')
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
import re
```

```python
dataset = '/content/E-Commerce Reviews.csv'
missing_cells = ['na', 'NA', 'N/A', 'n/a', '-', np.nan]
ecom = pd.read_csv(dataset, na_values=missing_cells)
ecom
```

```python
retain_cols = ['Review Text', 'Recommended IND']

for col in ecom.columns:
  if col not in retain_cols:
    ecom.drop(col, axis=1, inplace=True)
```

```python
ecom.isnull().sum()
```

```python
ecom.dropna(subset=['Review Text'], axis=0, how='any', inplace=True)
```

```python
ecom.duplicated().sum()
```

```python
ecom.drop_duplicates(inplace=True)
```

```python
ecom.shape
```

```python
ecom.rename(columns={'Review Text': 'Review_Text', 'Recommended IND': 'Recommended'}, inplace=True)
```

```python
lemmatization = WordNetLemmatizer()
stemmer = PorterStemmer()

result_lemmatiz = []
result_stemming = []
```

# NLP Pre-processing functions

```python
def process_text_lemmatiz(df, col, lemmatizer_obj, text_list):
  for index, row in tqdm(df.iterrows()):
    sentence = row[col]
    sentence = re.sub('[^a-zA-Z]', ' ', sentence)
    sentence = sentence.lower()
    sentence = sentence.split()
    sentence = [lemmatizer_obj.lemmatize(word) for word in sentence if word not in set(stopwords.words('english'))]
    sentence = ' '.join(sentence)
    text_list.append(sentence)
  return text_list


def process_text_stemming(df, col, stemmer_obj, text_list):
  for index, row in tqdm(df.iterrows()):
    sentence = row[col]
    sentence = re.sub('[^a-zA-Z]', ' ', sentence)
    sentence = sentence.lower()
    sentence = sentence.split()
    sentence = [stemmer_obj.stem(word) for word in sentence if word not in set(stopwords.words('english'))]
    sentence = ' '.join(sentence)
    text_list.append(sentence)
  return text_list
```

```python
process_text_lemmatiz(ecom, 'Review_Text', lemmatization, result_lemmatiz)
```

```python
print(len(result_lemmatiz), result_lemmatiz[22633])
```

```python
process_text_stemming(ecom, 'Review_Text', stemmer, result_stemming)
```
